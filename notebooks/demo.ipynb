```python
import bpp
import matplotlib.pyplot as plt

from bpp.catalog import create_uniform_catalog
from bpp.psf import get_gaussian_psf
from bpp.scene import create_scene
```

# Draw Scene

```python
# get astropy catalog of galaxy parameters
cat = create_uniform_catalog(n_rows=4, max_shift=4) # n_rows = how many galaxies?
```

```python
# create PSF galsim object
psf = get_gaussian_psf(fwhm=0.7)
```

```python
# draw the scene
pixel_scale = 0.2
slen = 81
scene = create_scene(slen=slen, catalog=cat, psf=psf, pixel_scale=pixel_scale,
                     g1=0.0, g2=0.0, sky_level=100, seed=0)
```

```python
# let's look at it
plt.figure(figsize=(7,7))
plt.imshow(scene)
plt.colorbar()

# with centers
for row in cat: 
    plt.scatter(row['ra']/ pixel_scale  - 0.5 + slen/2, row['dec'] / pixel_scale - 0.5 + slen/2, marker='x', color='r')
```
